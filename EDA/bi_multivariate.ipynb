# heatmap(categorical-categorical)

```python
sns.heatmap(pd.crosstab(titanic["Pclass"],titanic["Survived"]))
```

# cluster map(categorical- categorical)

```python
sns.clustermap(pd.crosstab(titanic["SibSp"],titanic["Survived"]))
```

# line plot (numerical-numerical)
- use it when we have time based quantity like year and place it at x-axis

```python
flights.head()
```

```python
flights['year'] = pd.to_numeric(flights['year'], errors='coerce')

new = flights.groupby("year").sum(numeric_only=True).reset_index()
```

```python
sns.lineplot(data=new,x="year",y="passengers")
```
