```python
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv(r"D:\machine learning\linear regression\simple linear regression\advertising.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
X = df[['TV']]
Y= df[['Sales']]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
# applying simple linear regression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
Y_pred=lr.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mae = mean_absolute_error(Y_pred,Y_test)
mae
```

```python
mse = mean_squared_error(Y_pred,Y_test)
mse
```

```python
import math
root_mse=math.sqrt(mse)
root_mse
```

```python
r2_s = r2_score(Y_pred,Y_test)
r2_s
```

```python
plt.scatter(X,Y)
plt.plot(X, lr.predict(X), color='red', label='Regression Line')
plt.xlabel('TV Advertising Budget (in $1000s)')
plt.ylabel('Sales (in $1000s)')
plt.legend()
plt.show()
```
